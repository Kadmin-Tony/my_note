```python
import pandas as pd
import numpy as np
from random import randint
```

```python
a = pd.DataFrame(np.array([randint(1, 100) for i in range(100)]).reshape(10, 10))
a
```

# 查询

```python
a.head(2)  # 查询前两排
```

```python
a.tail(3)  # 查询后三排
```

```python
a[2:7]  # 带冒号及为取行
```

```python
a[3]  # 仅取列标为3的那一列
```

```python
a[[4, 7]]  # 取列标为4与7的列
```

```python
a.loc[3, 4]  # 取第三行四列的值（通过标签进行搜索）
a.iloc[3, 4]  # 同上（通过位置进行搜索）
"""两者均可通过[:,:]进行连续搜索"""
```

# 查看统计信息
&nbsp;&nbsp; count: 数量 </br>
&nbsp;&nbsp; mean: 平均值 </br>
&nbsp;&nbsp; std: 方差 </br>

```python
a.mean()
```

```python
a.describe()
```

```python
a.sort_values(by=3)  # 按照第三列进行排序
```
